```python
import numpy as np
import pygimli as pg
import matplotlib.pyplot as plt

import models
```

#### Load Sample file

```python
data_loaded = np.loadtxt("sample.txt",  dtype=float, delimiter=",")

times = data_loaded[:,1]
eta = data_loaded[:,2]
```

```python
times
```

```python
eta
```

#### Fit

```python
model, response = models.fitCCTD(times, eta)

eta0, tau, c = model
eta_resp = response
```

#### Plot

```python
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(10, 8))
plt.plot(times, eta*1000,'-b+', label="data")
plt.plot(times, eta_resp*1000,'-r', linewidth=1, label=f"CC")
plt.xlabel('Time [sec]',fontsize = 18)
plt.ylabel('Chargeability [mV/V]',fontsize = 18)
plt.xlim([0, np.max(times)+0.2])
plt.ylim([0, np.max(eta)*1000+10])
plt.legend()
plt.grid(True)
plt.title(fr"CC $\eta_0$={eta0:.3f}, $\tau$={tau:.3f} c={c:.3f}")
plt.savefig("test1.png")
plt.show()  
```
